````python
import requests
from dotenv import load_dotenv
import os
import pandas as pd
import json

def search_books(query):   
    # query string 문자열을 dict 선언
    payload = {
    'query':  query, #'파이썬',
    'display': 50,
    'sort': 'sim'
    }

    # .env 파일에서 환경 변수 로드
    load_dotenv()

    # 환경 변수에서 값 읽기
    client_id = os.getenv("NAVER_CLIENT_ID")  # .env 파일의 NAVER_CLIENT_ID
    client_secret = os.getenv("NAVER_CLIENT_SECRET")  # .env 파일의 NAVER_CLIENT_SECRET

    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }
    url = 'https://openapi.naver.com/v1/search/book.json'
    
    # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)
    # json() 함수로 응답 결과 가져오기
    items_data = res.json()['items']
    
    # 1. 질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기 data/books.json 파일로 저장해 주세요.
    if not os.path.exists('data'):
        os.makedirs('data')
    with open('data/books.json', 'w', encoding='utf-8') as f:
        # json.dumps() 함수로 json 파일로 저장하기
        f.write(json.dumps(items_data, ensure_ascii=False, indent=4))

    # 2. books.json 파일을 Pandas DataFrame로 저장하기
    df = pd.read_json('data/books.json')

    # 3. 질문 :  검색어로  찾은  책 목록 출력하기
    display(df)

    # 4. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
    # title,author,discount,publisher,pubdate 컬럼만 출력
    # 가격은 descending (내림차순), index 초기화
    df2 = df[df['discount'] >= 20000][['title','author','discount','publisher','pubdate']].sort_values(by='discount', ascending=False).reset_index(drop=True)
    display(df2)
    # 6. 질문 :  검색어로  찾은  책 목록 중에서 출판사가 인피니티북스인 책만 출력하기
    # image , description 컬럼은 제외한 모든 컬럼 출력하기
    #index 는 초기화
    df3 = df[df['publisher'] == '인피니티북스'].drop(columns=['image', 'description']).reset_index(drop=True)
    display(df3)

search_books('파이썬')
````

def search_shops(query):

    payload = {
    'query':  query, #'가디건',
    'display': 50,
    'sort': 'sim'
    }

4. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 특정 쇼핑몰 상품만 출력하기
lprice 부터 brand  컬럼까지 slicing 으로 선택하여 출력
가격은 ascending (오름차순), index 초기화


````python
import requests
from dotenv import load_dotenv
import os
import pandas as pd
import json

def search_shops(query):
    load_dotenv()

    # 환경 변수에서 값 읽기
    client_id = os.getenv("NAVER_CLIENT_ID")  # .env 파일의 NAVER_CLIENT_ID
    client_secret = os.getenv("NAVER_CLIENT_SECRET")  # .env 파일의 NAVER_CLIENT_SECRET

    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret,
    }
    
    payload = {
    'query':  query, #'가디건',
    'display': 50,
    'sort': 'sim'
    }
    url = 'https://openapi.naver.com/v1/search/shop.json'
   
   # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)
    res.raise_for_status()  # HTTPError 발생 시 예외 처리
    items_data = res.json()['items']
    if not os.path.exists('data'):
        os.makedirs('data')
    with open('data/shops.json', 'w', encoding='utf-8') as f:
        # json.dumps() 함수로 json 파일로 저장하기
        f.write(json.dumps(items_data, ensure_ascii=False, indent=4))
    # 2. shops.json 파일을 Pandas DataFrame로 저장하기
    df = pd.read_json('data/shops.json')
    # 3. 질문 :  검색어로  찾은  Shop의 상품  목록 출력하기
    display(df)

    # 4. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 가격이 50,000원 이하인 상품만 출력하기
    # brand,lprice,mallName,link 컬럼만 출력
    # 가격은 ascending (오름차순), index 초기화
    df2 = df[df['lprice'] <= 50000][['brand','lprice','mallName','link']].sort_values(by='lprice', ascending=True).reset_index(drop=True)
    display(df2)

    # 4. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 특정 쇼핑몰 상품만 출력하기
    # lprice 부터 brand  컬럼까지 slicing 으로 선택하여 출력
    # 가격은 ascending (오름차순), index 초기화
    df3 = df[df['mallName']=='네이버'].loc[:,'lprice':'brand'].sort_values(by='lprice', ascending=True).reset_index(drop=True)
    display(df3)

search_shops('가디건')
````
